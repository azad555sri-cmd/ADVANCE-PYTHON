# Nested While Loop

```python
i = 1
while i <=5:
    print('data science')
    j = 1
    while j<=4:
        print('technology')
        j = j+1

    i = i+1
    print()
```

```python
i = 1
while i <=5:
    print('data science', end =" ")
    j = 1
    while j<=4:
        print('technology', end =" ")
        j = j+1

    i = i+1
    print()
```

```python
i = 1
while i<=2:
    j=0
    while j<=2:
        print(i*j, end=" ")
        j += 1
        print()
        i += 1
```

```python
i = 1
while i<=2 :
    j=0
    while j <=2 :
        print(i*j, end=" ")
        j += 1
    print()
    i += 1
```

```python
i = 1
while i<=4 :
    j=0
    while j <=3 :
        print(i*j, end=" ")
        j += 1
    print()
    i += 1
```

```python
a1=[1,2,3]
a2=[4,5,6]

i = 0
while i < len(a1):
    j = 0
    while j < len(a2):
        print(a1[i], a2[j])
        j += 1
    print()
    i += 1
```
